```python
X=data.iloc[:,[0,1,2,3]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('THE ELBOW METHOD',fontsize = 15)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=4)
Y_km=kmeans.fit_predict(X)
Y_km
```

```python
plt.scatter(X[Y_km == 0, 0], X[Y_km == 0, 1],s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[Y_km == 1, 0], X[Y_km == 1, 1],s = 100, c = 'yellow', label = 'Iris-versicolour')
plt.scatter(X[Y_km == 2, 0], X[Y_km == 2, 1],s = 100, c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[: ,1],s=100,marker= 'o',color='black',label = 'Centroids')
plt.title('CLUSTURES OF IRIS', fontsize = 16)

plt.legend()
```
